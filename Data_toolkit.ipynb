### **THEORY QUESTIONS**
### **1. What is NumPy, and why is it widely used in Python?**

**Answer:**
NumPy is a Python library used for numerical and mathematical operations. It provides a powerful array object which is faster and more efficient than Python lists. NumPy is widely used because it supports large datasets, mathematical functions, and improves performance in data processing.


### **2. How does broadcasting work in NumPy?**

**Answer:**
Broadcasting allows NumPy to perform operations on arrays of different shapes. Instead of copying data, NumPy automatically expands the smaller array to match the larger one so that element-wise operations can be performed.



### **3. What is a Pandas DataFrame?**

**Answer:**
A Pandas DataFrame is a two-dimensional data structure with rows and columns. It is similar to an Excel spreadsheet and is used to store, manipulate, and analyze data efficiently.



### **4. Explain the use of the `groupby()` method in Pandas.**

**Answer:**
The `groupby()` method is used to group data based on a specific column. After grouping, aggregate functions like sum, mean, or count can be applied to each group.



### **5. Why is Seaborn preferred for statistical visualizations?**

**Answer:**
Seaborn is preferred because it provides attractive and informative statistical plots with minimal code. It works well with Pandas DataFrames and automatically handles statistical calculations.



### **6. What are the differences between NumPy arrays and Python lists?**

**Answer:**
NumPy arrays are faster, consume less memory, and support vectorized operations. Python lists are slower and can store different data types but are less efficient for numerical computation.



### **7. What is a heatmap, and when should it be used?**

**Answer:**
A heatmap is a graphical representation of data where values are shown using colors. It is mainly used to visualize correlations or patterns in large datasets.



### **8. What does the term “vectorized operation” mean in NumPy?**

**Answer:**
Vectorized operations allow NumPy to perform calculations on entire arrays without using loops. This makes the operations faster and the code simpler.



### **9. How does Matplotlib differ from Plotly?**

**Answer:**
Matplotlib is used to create static plots, while Plotly is used to create interactive plots. Plotly allows zooming, hovering, and better user interaction.


### **10. What is the significance of hierarchical indexing in Pandas?**

**Answer:**
Hierarchical indexing allows multiple levels of indexing in a DataFrame. It helps in organizing complex datasets and performing advanced data analysis.


### **11. What is the role of Seaborn’s `pairplot()` function?**

**Answer:**
The `pairplot()` function is used to visualize relationships between multiple numerical variables in a dataset through scatter plots and histograms.


### **12. What is the purpose of the `describe()` function in Pandas?**

**Answer:**
The `describe()` function provides a statistical summary of data such as count, mean, minimum, maximum, and standard deviation.



### **13. Why is handling missing data important in Pandas?**

**Answer:**
Handling missing data is important because missing values can affect the accuracy of analysis and produce incorrect results.



### **14. What are the benefits of using Plotly for data visualization?**

**Answer:**
Plotly provides interactive charts, easy sharing, web-based visualization, and better user engagement compared to static plots.



### **15. How does NumPy handle multidimensional arrays?**

**Answer:**
NumPy supports multidimensional arrays using the `ndarray` object. These arrays can represent data in 1D, 2D, or higher dimensions.



### **16. What is the role of Bokeh in data visualization?**

**Answer:**
Bokeh is used to create interactive and web-ready visualizations. It is mainly used for dashboards and browser-based applications.



### **17. Explain the difference between `apply()` and `map()` in Pandas.**

**Answer:**
`map()` works only on Pandas Series, while `apply()` works on both Series and DataFrames. `apply()` is more flexible.



### **18. What are some advanced features of NumPy?**

**Answer:**
Advanced features of NumPy include broadcasting, vectorization, linear algebra operations, Fourier transforms, and random number generation.


### **19. How does Pandas simplify time series analysis?**

**Answer:**
Pandas provides date-time indexing, resampling, rolling windows, and time-based filtering, which makes time series analysis easier.


### **20. What is the role of a pivot table in Pandas?**

**Answer:**
A pivot table is used to summarize large datasets by grouping and aggregating values in a structured format.



### **21. Why is NumPy’s array slicing faster than Python’s list slicing?**

**Answer:**
NumPy slicing creates views instead of copies, which reduces memory usage and increases speed.



### **22. What are some common use cases for Seaborn?**

**Answer:**
Seaborn is commonly used for correlation analysis, distribution plots, regression analysis, and exploratory data analysis.

```python
# Q4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?
import seaborn as sns
df = pd.DataFrame({"Maths": [80, 90, 70],
                   "Science": [85, 88, 75]})
sns.heatmap(df.corr(), annot=True)
```
